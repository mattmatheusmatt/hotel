## 3.2 Temporalidade

```python
def bar_count(hue,ano,dados,palette):
    list = ['January','February','March','April','May','June','July','August','September','October','November','December']
    sns.catplot(x='ArrivalDateMonth',
            hue=hue,
            kind='count',
            aspect= 12/6,
            data=dados[dados['ArrivalDateYear'] == ano],
            palette=palette,
            order=list)
```

```python
def bar_count_iscanceled(hue,ano,dados,palette,iscanceled):
    list = ['January','February','March','April','May','June','July','August','September','October','November','December']
    sns.catplot(x='ArrivalDateMonth',
            hue=hue,
            kind='count',
            aspect= 12/6,
            data=dados[(dados['ArrivalDateYear'] == ano) & (dados['IsCanceled'] == iscanceled)],
            palette=palette,
            order=list)
```

```python
bar_count(hue = 'IsCanceled' ,ano = 2015,dados = resort_hotel,palette = "Set2")
bar_count(hue = 'IsCanceled' ,ano = 2016,dados = resort_hotel,palette = "Set2")
bar_count(hue = 'IsCanceled' ,ano = 2017,dados = resort_hotel,palette = "Set2")
```

```python
bar_count(hue = 'IsCanceled' ,ano = 2015,dados = city_hotel,palette = "Set2")
bar_count(hue = 'IsCanceled' ,ano = 2016,dados = city_hotel,palette = "Set2")
bar_count(hue = 'IsCanceled' ,ano = 2017,dados = city_hotel,palette = "Set2")
```
